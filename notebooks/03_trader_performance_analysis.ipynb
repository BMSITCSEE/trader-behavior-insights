# Trader Performance Analysis

This notebook analyzes individual trader performance, identifies top performers, and segments traders based on behavior patterns.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import plotly.express as px
import plotly.graph_objects as go
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load data
df = pd.read_csv('../data/processed/merged_data.csv')
df['time'] = pd.to_datetime(df['time'])

print(f"Total trades: {len(df):,}")
print(f"Unique traders: {df['account'].nunique():,}")
print(f"Unique symbols: {df['symbol'].nunique():,}")
```

## 1. Individual Trader Performance Metrics

```python
# Calculate comprehensive trader metrics
trader_metrics = df.groupby('account').agg({
    'closedPnL': ['sum', 'mean', 'std', 'min', 'max'],
    'is_profitable': ['mean', 'sum'],
    'size': ['mean', 'sum'],
    'leverage': ['mean', 'max'],
    'trade_value': ['mean', 'sum'],
    'symbol': 'nunique',
    'time': ['min', 'max', 'count']
})

# Flatten column names
trader_metrics.columns = ['_'.join(col).strip() for col in trader_metrics.columns.values]
trader_metrics.columns = ['total_pnl', 'avg_pnl', 'pnl_std', 'min_pnl', 'max_pnl',
                          'win_rate', 'winning_trades', 'avg_size', 'total_volume',
                          'avg_leverage', 'max_leverage', 'avg_trade_value', 'total_trade_value',
                          'symbols_traded', 'first_trade', 'last_trade', 'total_trades']

# Calculate additional metrics
trader_metrics['profit_factor'] = df[df['closedPnL'] > 0].groupby('account')['closedPnL'].sum() / abs(df[df['closedPnL'] < 0].groupby('account')['closedPnL'].sum())
trader_metrics['sharpe_ratio'] = trader_metrics['avg_pnl'] / trader_metrics['pnl_std']
trader_metrics['trading_days'] = (pd.to_datetime(trader_metrics['last_trade']) - pd.to_datetime(trader_metrics['first_trade'])).dt.days + 1
trader_metrics['avg_trades_per_day'] = trader_metrics['total_trades'] / trader_metrics['trading_days']

# Fill inf values
trader_metrics = trader_metrics.replace([np.inf, -np.inf], np.nan)

print(f"Trader metrics calculated for {len(trader_metrics)} traders")
trader_metrics.head()
```

## 2. Top Performers Analysis

```python
# Identify top performers by different metrics
top_by_pnl = trader_metrics.nlargest(10, 'total_pnl')[['total_pnl', 'win_rate', 'total_trades', 'sharpe_ratio']]
top_by_winrate = trader_metrics[trader_metrics['total_trades'] >= 50].nlargest(10, 'win_rate')[['win_rate', 'total_pnl', 'total_trades', 'avg_pnl']]
top_by_sharpe = trader_metrics[trader_metrics['total_trades'] >= 50].nlargest(10, 'sharpe_ratio')[['sharpe_ratio', 'total_pnl', 'win_rate', 'total_trades']]

print("\n=== TOP PERFORMERS ===")
print("\nTop 10 by Total PnL:")
print(top_by_pnl)
print("\nTop 10 by Win Rate (min 50 trades):")
print(top_by_winrate)
print("\nTop 10 by Sharpe Ratio (min 50 trades):")
print(top_by_sharpe)
```

```python
# Visualize top performers
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# PnL distribution
ax1 = axes[0, 0]
trader_metrics['total_pnl'].hist(bins=50, ax=ax1, edgecolor='black')
ax1.axvline(trader_metrics['total_pnl'].median(), color='red', linestyle='--', label='Median')
ax1.set_title('Distribution of Total PnL')
ax1.set_xlabel('Total PnL')
ax1.set_ylabel('Number of Traders')
ax1.legend()

# Win rate distribution
ax2 = axes[0, 1]
trader_metrics['win_rate'].hist(bins=30, ax=ax2, edgecolor='black')
ax2.axvline(trader_metrics['win_rate'].mean(), color='red', linestyle='--', label='Mean')
ax2.set_title('Distribution of Win Rates')
ax2.set_xlabel('Win Rate')
ax2.set_ylabel('Number of Traders')
ax2.legend()

# Scatter: Total PnL vs Win Rate
ax3 = axes[1, 0]
scatter = ax3.scatter(trader_metrics['win_rate'], trader_metrics['total_pnl'], 
                     c=trader_metrics['total_trades'], c
                          cmap='viridis', alpha=0.6)
ax3.set_xlabel('Win Rate')
ax3.set_ylabel('Total PnL')
ax3.set_title('Total PnL vs Win Rate')
cbar = plt.colorbar(scatter, ax=ax3)
cbar.set_label('Total Trades')

# Trading activity by profit tier
ax4 = axes[1, 1]
trader_metrics['profit_tier'] = pd.qcut(trader_metrics['total_pnl'], q=5, 
                                        labels=['Bottom 20%', 'Low', 'Mid', 'High', 'Top 20%'])
profit_tiers = trader_metrics.groupby('profit_tier').agg({
    'total_trades': 'mean',
    'avg_leverage': 'mean',
    'symbols_traded': 'mean'
})
profit_tiers.plot(kind='bar', ax=ax4)
ax4.set_title('Trading Characteristics by Profit Tier')
ax4.set_xlabel('Profit Tier')
ax4.set_ylabel('Average Value')
ax4.legend(['Avg Trades', 'Avg Leverage', 'Avg Symbols'])

plt.tight_layout()
plt.savefig('../results/figures/trader_performance_overview.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 3. Trader Segmentation using Machine Learning

```python
# Prepare features for clustering
clustering_features = ['total_pnl', 'win_rate', 'avg_leverage', 'total_trades', 
                      'avg_trade_value', 'symbols_traded', 'sharpe_ratio']

# Filter traders with minimum activity
active_traders = trader_metrics[trader_metrics['total_trades'] >= 10].copy()
print(f"Analyzing {len(active_traders)} active traders (10+ trades)")

# Handle missing values
X = active_traders[clustering_features].fillna(0)

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine optimal number of clusters
silhouette_scores = []
K_range = range(2, 10)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    silhouette_scores.append(score)

# Plot silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(K_range, silhouette_scores, 'bo-')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Optimal Number of Clusters')
plt.grid(True)
plt.show()

# Use optimal number of clusters
optimal_k = K_range[np.argmax(silhouette_scores)]
print(f"Optimal number of clusters: {optimal_k}")
```

```python
# Perform final clustering
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)  # Using 4 clusters for clear segmentation
active_traders['cluster'] = kmeans.fit_predict(X_scaled)

# Analyze clusters
cluster_summary = active_traders.groupby('cluster').agg({
    'total_pnl': ['mean', 'median'],
    'win_rate': 'mean',
    'avg_leverage': 'mean',
    'total_trades': 'mean',
    'sharpe_ratio': 'mean',
    'symbols_traded': 'mean'
}).round(2)

cluster_summary.columns = ['_'.join(col).strip() for col in cluster_summary.columns.values]

# Add cluster size
cluster_summary['cluster_size'] = active_traders['cluster'].value_counts().sort_index()

print("\nCluster Analysis:")
print(cluster_summary)

# Assign meaningful names to clusters based on characteristics
cluster_names = {
    0: 'Conservative Traders',
    1: 'High Volume Traders',
    2: 'Risk Takers',
    3: 'Professional Traders'
}

active_traders['trader_type'] = active_traders['cluster'].map(cluster_names)
```

```python
# Visualize clusters using PCA
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)

# Create 3D scatter plot
fig = px.scatter_3d(
    x=X_pca[:, 0], y=X_pca[:, 1], z=X_pca[:, 2],
    color=active_traders['trader_type'],
    title='Trader Clusters in 3D Space (PCA)',
    labels={'x': 'PC1', 'y': 'PC2', 'z': 'PC3'},
    hover_data={'Account': active_traders.index,
                'Total PnL': active_traders['total_pnl'].round(2),
                'Win Rate': active_traders['win_rate'].round(3)}
)

fig.update_layout(height=700)
fig.show()

# Feature importance
feature_importance = pd.DataFrame({
    'feature': clustering_features,
    'PC1': pca.components_[0],
    'PC2': pca.components_[1],
    'PC3': pca.components_[2]
})

print("\nPCA Feature Contributions:")
print(feature_importance)
```

## 4. Trading Pattern Analysis

```python
# Analyze trading patterns for each cluster
cluster_patterns = {}

for cluster_name in cluster_names.values():
    cluster_traders = active_traders[active_traders['trader_type'] == cluster_name].index
    cluster_trades = df[df['account'].isin(cluster_traders)]
    
    patterns = {
        'preferred_hours': cluster_trades.groupby('hour').size().idxmax(),
        'preferred_symbols': cluster_trades['symbol'].value_counts().head(3).to_dict(),
        'avg_holding_time': 'N/A',  # Would need entry/exit pairs
        'sentiment_preference': cluster_trades['Classification'].value_counts().to_dict(),
        'monthly_pattern': cluster_trades.groupby(cluster_trades['time'].dt.month).size().to_dict()
    }
    
    cluster_patterns[cluster_name] = patterns

# Create visualization of trading patterns
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

for idx, (cluster_name, ax) in enumerate(zip(cluster_names.values(), axes.flat)):
    cluster_traders = active_traders[active_traders['trader_type'] == cluster_name].index
    cluster_trades = df[df['account'].isin(cluster_traders)]
    
    # Hourly trading pattern
    hourly_pattern = cluster_trades.groupby('hour').size()
    ax.bar(hourly_pattern.index, hourly_pattern.values, alpha=0.7)
    ax.set_title(f'{cluster_name} - Hourly Trading Pattern')
    ax.set_xlabel('Hour of Day')
    ax.set_ylabel('Number of Trades')
    ax.set_xlim(-0.5, 23.5)

plt.tight_layout()
plt.savefig('../results/figures/cluster_trading_patterns.png', dpi=300, bbox_inches='tight')
plt.show()
```

## 5. Performance Evolution Over Time

```python
# Track performance evolution for top traders
top_traders = trader_metrics.nlargest(20, 'total_pnl').index

# Calculate cumulative PnL over time
performance_evolution = []

for trader in top_traders:
    trader_data = df[df['account'] == trader].sort_values('time')
    trader_data['cumulative_pnl'] = trader_data['closedPnL'].cumsum()
    trader_data['trader'] = trader
    performance_evolution.append(trader_data[['time', 'cumulative_pnl', 'trader']])

evolution_df = pd.concat(performance_evolution)

# Interactive plot
fig = px.line(evolution_df, x='time', y='cumulative_pnl', color='trader',
              title='Performance Evolution of Top 20 Traders',
              labels={'cumulative_pnl': 'Cumulative PnL', 'time': 'Date'})
fig.update_layout(height=600, showlegend=False)
fig.show()
```

## 6. Key Findings and Trader Profiles

```python
# Generate trader profiles
trader_profiles = {}

for cluster_id, cluster_name in cluster_names.items():
    cluster_data = active_traders[active_traders['cluster'] == cluster_id]
    
    profile = {
        'name': cluster_name,
        'size': len(cluster_data),
        'characteristics': {
            'avg_total_pnl': cluster_data['total_pnl'].mean(),
            'avg_win_rate': cluster_data['win_rate'].mean(),
            'avg_leverage': cluster_data['avg_leverage'].mean(),
            'avg_trades': cluster_data['total_trades'].mean(),
            'avg_sharpe': cluster_data['sharpe_ratio'].mean()
        },
        'top_performer': cluster_data['total_pnl'].idxmax(),
        'trading_style': 'Aggressive' if cluster_data['avg_leverage'].mean() > 5 else 'Conservative',
        'risk_level': 'High' if cluster_data['pnl_std'].mean() > trader_metrics['pnl_std'].median() else 'Low'
    }
    
    trader_profiles[cluster_name] = profile

# Display profiles
for name, profile in trader_profiles.items():
    print(f"\n{'='*50}")
    print(f"TRADER PROFILE: {name}")
    print(f"{'='*50}")
    print(f"Number of traders: {profile['size']}")
    print(f"Trading style: {profile['trading_style']}")
    print(f"Risk level: {profile['risk_level']}")
    print(f"\nAverage Metrics:")
    for metric, value in profile['characteristics'].items():
        print(f"  - {metric}: {value:.3f}")
    print(f"\nTop performer: {profile['top_performer']}")

# Save profiles
import json
with open('../results/reports/trader_profiles.json', 'w') as f:
    json.dump(trader_profiles, f, indent=4, default=str)
```
